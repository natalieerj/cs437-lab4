```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df0 = pd.read_csv('vehicle0.csv')
df1 = pd.read_csv('vehicle1.csv')
df2 = pd.read_csv('vehicle2.csv')
df3 = pd.read_csv('vehicle3.csv')
df4 = pd.read_csv('vehicle4.csv')
```

```python
df0
```

```python
df0['co2_cumsum'] = df0['vehicle_CO2'].cumsum()
df1['co2_cumsum'] = df1['vehicle_CO2'].cumsum()
df2['co2_cumsum'] = df2['vehicle_CO2'].cumsum()
df3['co2_cumsum'] = df3['vehicle_CO2'].cumsum()
df4['co2_cumsum'] = df4['vehicle_CO2'].cumsum()
df0['helpercol'] = (df0['vehicle_x']**2 +df0['vehicle_y']**2 )**.5
df0['cumDist'] = df0['helpercol'].cumsum()
df1['helpercol'] = (df1['vehicle_x']**2 +df1['vehicle_y']**2 )**.5
df1['cumDist'] = df1['helpercol'].cumsum()
df2['helpercol'] = (df2['vehicle_x']**2 +df2['vehicle_y']**2 )**.5
df2['cumDist'] = df2['helpercol'].cumsum()
df3['helpercol'] = (df3['vehicle_x']**2 +df3['vehicle_y']**2 )**.5
df3['cumDist'] = df3['helpercol'].cumsum()
df4['helpercol'] = (df4['vehicle_x']**2 +df4['vehicle_y']**2 )**.5
df4['cumDist'] = df4['helpercol'].cumsum()
```
